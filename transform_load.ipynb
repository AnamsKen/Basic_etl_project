```python
import pandas as pd
import numpy as np
import psycopg2
from sqlalchemy import create_engine
```

```python
#Reading dataframe
df = pd.read_csv("C:/Users/user/Desktop/DE/Loading_data_project/jumia_oraimo_dataset_new.csv")
```

```python
df.head()
```

```python
#Cleaning and transforming dataset
def clean_data():
    df['old_price'] = df['old_price'].str.replace('₦ ','')
    df['discounted_price'] = df['discounted_price'].str.replace('₦ ','')
    df['discount'] = df['discount'].str.replace('%','')

clean_data()
```

```python
df.head()
```

```python
#Connecting to database
conn = create_engine('postgresql://postgres:Amampjdm@localhost:5432/ecommerce')
```

```python
#Loading to database
df.to_sql('jumia_oraimo', conn, if_exists= 'replace')
```
